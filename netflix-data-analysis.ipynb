```python
data.head()
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
content = data.groupby('year_added')['month_added'].value_counts().unstack().fillna(0)[month_order].T

plt.figure(figsize=(10,8))
plt.title("Yıllara göre aylık eklenen içerik sayısı")
sns.heatmap(content , cmap = 'Blues')
plt.show()
```
